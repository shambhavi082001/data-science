```python
x = np.arange(-10,10,0.1)
print(x.shape)
```

```python
x2 = x**2 
```

```python
plt.plot(x,x2)
```

```python
x3 = x ** 3
```

```python
plt.figure(figsize=(15,7))
plt.plot(x,label='original',linestyle='--')
plt.plot(x2,label='squared',linestyle=':')
plt.plot(x3,label='cubed')
plt.legend() #activate the legend
plt.xlabel("data points") #add a lavel to the x axis
plt.ylabel("value of x")  #add a lavel to the y axis
plt.title("the visualization of x and its square and cube") # add a title to the figure
plt.show() #show the figure
```

```python
cities = ['lucknow','delhi','mumbai','pune','banglore']
rainfall = [39,20,62,45,78]
```

```python
plt.bar(cities, rainfall)
```

```python
plt.barh(cities, rainfall)
```

```python
plt.figure(figsize=(10,7),facecolor='black')
plt.bar(cities,rainfall,color='#ff8800',edgecolor='black',hatch='//') #width=[.1,.2,.3,.4,.5],align='edge'
plt.title("rainfall comparison of some cities")
```

```python
plt.figure(facecolor='white')
plt.pie(rainfall, labels=cities, autopct='%1.1f%%', radius=2)
plt.show()
```

```python
plt.figure(facecolor='white')
wedges = [0,.3,0,0,0]
plt.pie(rainfall, labels=cities, autopct='%1.1f%%', radius=2,explode=wedges)
plt.show()
```
